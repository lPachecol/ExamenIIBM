```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, y):
    return -5*y + 5*t**2 + 2*t

def exact_solution(t):
    return t**2 + (1/3)*np.exp(-5*t)

def euler(f, t0, y0, h, N):
    t_values = [t0]
    y_values = [y0]

    for i in range(N):
        t = t_values[-1]
        y = y_values[-1]
        y_next = y + h * f(t, y)
        t_next = t + h
        t_values.append(t_next)
        y_values.append(y_next)

    return np.array(t_values), np.array(y_values)

t0 = 0
tf = 1
y0 = 1/3
N = 10
h = (tf - t0) / N

t_vals, y_approx = euler(f, t0, y0, h, N)

y_exact = exact_solution(t_vals)

error_relativo = np.abs((y_exact - y_approx) / y_exact)
error_promedio = np.mean(error_relativo)
print("Error promedio relativo:", error_promedio)

plt.plot(t_vals, y_approx, label="Euler", marker='o')
plt.plot(t_vals, y_exact, label="Solución exacta", linestyle='--')
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Comparación entre Euler y la solución exacta")
plt.legend()
plt.grid(True)
plt.show()
```
