```python
import numpy as np
import matplotlib.pyplot as plt

def S0(x):
    return -4*(x + 1)**3 + 3*(x + 1)**2 + 6*(x + 1) + 1

def S1(x):
    return x**3 - 3*x**2 + 3*x + 5

x0 = np.linspace(-1, 0, 100)
x1 = np.linspace(0, 1, 100)

y0 = S0(x0)
y1 = S1(x1)

x_data = np.array([-1, 0, 1])
y_data = np.array([1, 5, 3])

x_tangent = np.linspace(-0.5, 0.5, 10)
y_tangent = -3 * (x_tangent - 0) + 5  

plt.figure(figsize=(8, 5))
plt.plot(x0, y0, 'r', label="Spline S0 ([-1, 0])")
plt.plot(x1, y1, 'b', label="Spline S1 ([0, 1])")
plt.plot(x_tangent, y_tangent, 'k--', label="Pendiente deseada m=-3")
plt.plot(x_data, y_data, 'ko', label="Puntos dados")

plt.scatter(0, 5, s=80, color='cyan', edgecolors='black', zorder=5, label="Pendiente forzada")
plt.title("Spline cbico con pendiente forzada en (0, 5)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
